```python
#EDGE DETECTION
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
from skimage.draw import polygon

# -------------------------------------------------
# 1. Train MLP using XOR logic
# -------------------------------------------------

X = np.array([
[0,0],
[0,1],
[1,0],
[1,1]
])

y = np.array([0,1,1,0])

mlp = MLPClassifier(hidden_layer_sizes=(2,),
                    activation='logistic',
                    solver='lbfgs',
                    max_iter=5000)

mlp.fit(X,y)

# -------------------------------------------------
# 2. Create Binary Image
# -------------------------------------------------

img = np.zeros((120,240))

# Rectangle
img[40:80,20:70] = 1

# Triangle
for i in range(40):
    img[40+i,120-i:120+i] = 1

# -------------------------------------------------
# 3. Generate Proper 5-Point Star
# -------------------------------------------------

cx, cy = 60, 200   # star center
R = 25             # outer radius
r = 10             # inner radius

points = []

for i in range(10):

    angle = i * np.pi / 5

    if i % 2 == 0:
        x = cx + R * np.sin(angle)
        y = cy + R * np.cos(angle)
    else:
        x = cx + r * np.sin(angle)
        y = cy + r * np.cos(angle)

    points.append((x,y))

points = np.array(points)

rr, cc = polygon(points[:,0], points[:,1])

img[rr.astype(int), cc.astype(int)] = 1

# -------------------------------------------------
# 4. Edge Detection using XOR MLP
# -------------------------------------------------

edges = np.zeros_like(img)

rows,cols = img.shape

for i in range(1,rows-1):
    for j in range(1,cols-1):

        p = img[i][j]

        neighbors = [
            img[i][j+1],
            img[i][j-1],
            img[i+1][j],
            img[i-1][j]
        ]

        for n in neighbors:
            edge = mlp.predict([[p,n]])[0]
            if edge == 1:
                edges[i][j] = 1
                break

# -------------------------------------------------
# 5. Display Images
# -------------------------------------------------

plt.figure(figsize=(16,6))

plt.subplot(1,2,1)
plt.title("Binary Image")
plt.imshow(img, cmap='gray', vmin=0, vmax=1, interpolation='nearest')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Edge Detected Image")
plt.imshow(edges, cmap='gray', vmin=0, vmax=1, interpolation='nearest')
plt.axis('off')

plt.show()
```

```python

import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns



beer_data = pd.read_csv("/root/.cache/kagglehub/datasets/avd1729/beer-csv/versions/1/beer.csv")
beer_data.head(5)
```

```python

scaler = StandardScaler()
scaled_beer = scaler.fit_transform(beer_data[['calories','sodium','alcohol','cost']])
scaled_beer
```

```python

kmeans = KMeans(n_clusters=3, random_state=0, n_init='auto')
beer_data['cluster'] = kmeans.fit_predict(scaled_beer)
beer_data.head()
```

```python
for i in range(kmeans.n_clusters):
    print(f"Cluster {i}:")
    display(beer_data[beer_data['cluster'] == i])
    print("\n")
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(data=beer_data, x='alcohol', y='calories', hue='cluster', palette='viridis', s=100)
plt.title('KMeans Clustering of Beer Data (Alcohol vs. Calories)')
plt.xlabel('Alcohol Content')
plt.ylabel('Calories')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
#Multilayer perceptron
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPClassifier
import numpy as np


np.random.seed(1)
n_sample = 30
x = np.random.rand(n_sample, 2)
x_binary = (x >= 0.5).astype(int)
y = np.logical_xor(x_binary[:, 0], x_binary[:, 1]).astype(int)


mlp = MLPClassifier(hidden_layer_sizes=(2,),
                    activation='tanh',
                    solver='adam',
                    learning_rate_init=0.1,
                    max_iter=5000,
                    random_state=1
)

mlp.fit(x,y)
xx,yy = np.meshgrid(np.linspace(0,1,200),np.linspace(0,1,200))

z = mlp.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)


plt.contourf(xx,yy,z,alpha=0.3)
plt.scatter(x[:, 0],x[:, 1],c=y, edgecolors='k')

plt.axvline(0.5,linestyle='--',color='b')
plt.axhline(0.5,linestyle='--',color='b')

plt.title('XOR Decision Boundary')
plt.xlabel('X1')
plt.ylabel('X2')
plt.show()
```
