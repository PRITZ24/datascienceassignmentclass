```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# -----------------------------
# Generate highly accurate data
# -----------------------------
np.random.seed(7)

# Large sample size for smoother histogram
data = np.random.normal(loc=0, scale=1, size=500000)

# -----------------------------
# Figure setup
# -----------------------------
fig, ax = plt.subplots(figsize=(14, 8))

# -----------------------------
# Histogram
# -----------------------------
ax.hist(
    data,
    bins=120,
    density=True,
    color='#2F6DB5',
    alpha=0.82,
    edgecolor='white',
    linewidth=0.35,
    label='Empirical Data (Histogram)'
)

# -----------------------------
# Theoretical Normal Curve
# -----------------------------
x = np.linspace(-4, 4, 3000)
y = norm.pdf(x, 0, 1)

ax.plot(
    x,
    y,
    color='red',
    linewidth=3.2,
    label='Theoretical Normal PDF',
    zorder=5
)

# -----------------------------
# Styling
# -----------------------------
ax.set_title(
    'Normal Distribution',
    fontsize=30,
    pad=12,
    fontweight='bold'
)

ax.set_xlabel(
    'Value',
    fontsize=24
)

ax.set_ylabel(
    'Probability Density',
    fontsize=24
)

ax.set_xlim(-4, 4)
ax.set_ylim(0, 0.45)

# Grid
ax.grid(
    True,
    alpha=0.12,
    linestyle='-'
)

# Legend
ax.legend(
    fontsize=18,
    loc='upper right',
    framealpha=0.95
)

# Tick size
ax.tick_params(
    axis='both',
    labelsize=14
)

# Tight layout
plt.tight_layout()

# Display graph
plt.show()
```
